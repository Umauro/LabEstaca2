### 1- Simulación

* Corto: 6 min
* Medio: 8 min

4. Determine la probabilidad de que, en algun momento del trayecto el conductor elija al menos 5 veces consecutivas el camino Medio. Nota: si llega a suceder algo del estilo: m-m-m-m-m-c-m-m-m-m-m-l cuenta como 2 "éxitos".
